## Line plot | 线图

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# Note that using plt.subplots below is equivalent to using
# fig = plt.figure and then ax = fig.add_subplot(111)
fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()

fig.savefig("test.png")
plt.show()
```

我们简单分析一下这段代码：
- 导入模块
    - matplotlib是一个很大的模块，我们使用它的子模块pyplot来画图，关于pyplot用法后面细讲。
    - 绘图一般和numpy结合使用。numpy是一个出色的数学库，提供了高效的n维数组操作。MATLAB以其高效的矩阵操作闻名，通过numpy，Python也具有了MATLAB的这个优点。
- 生成绘图向量
    - np中arrange方法同Python的range方法类似。
    - np中提供了常用数学操作。
- 创建图像
    - fig为当前图像（figure）
    - ax为当前坐标图（axes）
- 装饰图像
    - ax.set()设置坐标轴，题目。
    - ax.grid()设置背景格
- 保存图像
    - 通过fig.savefig("xxx.png")保存图像
    - 只支持png格式。
- 显示图像
    - plt.show()显示图像，如果不加这句话，默认在脚本中不显示

## Subplot | 子图

```python
import numpy as np
import matplotlib.pyplot as plt


x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```

我们简单分析一下代码：
- subplot使用方法和MATLAB完全一致
- polt绘图时指定线段类型和MATLAB完全一致
- 给坐标轴加标签方法和MATLAB完全一致，注意到上个例子我们使用了axes对象的set方法，而这次直接使用了xlabel，ylabel，title方法。

## Histograms | 直方图

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

np.random.seed(19680801)

# example data
mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(437)

num_bins = 50

fig, ax = plt.subplots()

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, normed=1)

# add a 'best fit' line
y = mlab.normpdf(bins, mu, sigma)
ax.plot(bins, y, '--')
ax.set_xlabel('Smarts')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
```

我们简单分析一下代码，其实注释写的很详细了：
- 直方图有关的方法为hist()，具体方法后面再讲，不过API可读性非常强。
- 直方图和我们的pdf图重叠了，其实matplotlib中，同一个axes多个数据(lines)默认是重叠的，要关闭的话，使用hold(False)。
- 注意到：我们设置label和title的方法又换了一种，“官方炫技，最为致命”。

## 3D Surface | 三维表面图

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

## Scatter | 散点图

```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680801)


N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```
